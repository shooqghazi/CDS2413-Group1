```python
# Display a countplot to display education_level
sns.countplot(data=dataset, x="Education Level",color="orange")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=dataset,x="Stress Level",hue="Anxious Before Exams",palette="dark:r")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=dataset,x="Education Level",palette="dark:y",hue="Stress Level")
plt.xticks(rotation=45)
plt.show()
```

```python
# First, define the numeric_col variable with the column names you want to plot
# For example, if your dataset has columns like 'age', 'income', 'height', etc.
numeric_col = ['Age', 'Screen Time (hrs/day)', 'Sleep Duration (hrs)', 'Physical Activity (hrs/week)']  # Replace with your actual numeric column names

# Now the rest of your code will work
colors=['g', 'b', 'r', 'c', 'm', 'y', 'k']
for i, col in enumerate(numeric_col):
    sns.boxplot(x=dataset[col], color=colors[i%len(colors)])
    plt.title("Boxplot of {}".format(col))
    plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(dataset[["Age", "Screen Time (hrs/day)", "Sleep Duration (hrs)", "Physical Activity (hrs/week)"]].corr(), annot=True, cmap="RdPu")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
